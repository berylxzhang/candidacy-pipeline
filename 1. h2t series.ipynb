# 1. modules imported

```python
%matplotlib inline
import numpy as np
import mdtraj as md
import matplotlib.pyplot as plt
```

# 2. head-to-tail distance calculation

```python
traj = md.load('trj_whole.xtc', top='traj_pbc.gro')
```

```python
dist = md.compute_distances(traj,[[0,23]])
length = len(dist)
```

```python
fig, ax = plt.subplots(figsize=(15,6))
x = np.array(range(traj.xyz.shape[0]))
ax.plot(x/100, dist)
ax.set_xlabel("time, t(ns)",fontsize=18)
ax.set_ylabel("h2t distance, l ($\AA$)",fontsize=18) 
ax.tick_params(labelsize=15)
plt.draw()
plt.show()
plt.savefig('h2tdistance.png')
```

```python
np.savetxt('h2t series.txt',dist)
```

```python
print('the maximum h2t distance is %2.4f, the minimum h2t distance is %2.4f'%(max(dist),min(dist)))
```
